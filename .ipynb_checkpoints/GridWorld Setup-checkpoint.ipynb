# Gridworld Bayesian RL
This notebook aims to setup a standard and a more advanced gridworld problem for implementing a Bayesian Inference RL approach

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import os 
from matplotlib.patches import Rectangle, FancyArrow
from matplotlib.patches import Patch
from collections import defaultdict

```

```python
def plot_gridworld(gridworld,start_cells,forbidden_cells,end_cells,gradient,actions):
    
    #ticks = np.arange(0,10,1)
    x_ticks = np.arange(gridworld.shape[1]) 
    y_ticks = np.arange(gridworld.shape[0]) 

    if gradient:
        display_grid = gridworld
        cmap = 'viridis'
        vmin=np.min(gridworld)
        vmax=np.max(gridworld)
        #print(vmin,vmax)
    else:
        cmap = 'gray'
        vmin = 0
        vmax = 1
        display_grid = np.ones_like(gridworld)
        
    plt.imshow(display_grid, cmap=cmap, vmin=vmin, vmax=vmax, alpha=1)
    #plt.gca().invert_yaxis()  # Add this line
    plt.xticks(x_ticks, minor=False)
    plt.yticks(y_ticks, minor=False)
    plt.gca().set_xticks(np.arange(-0.5, gridworld.shape[1], 1), minor=True)
    plt.gca().set_yticks(np.arange(-0.5, gridworld.shape[0], 1), minor=True)
    
    plt.grid(which='minor', color='black', linewidth=1)

    if len(actions)==0:
        # Add text annotations with gridworld values
        for i in range(gridworld.shape[0]):
            for j in range(gridworld.shape[1]):
                # do not print text if cell is starting or ending or forbidden
                    if ((i,j) not in (start_cells+forbidden_cells+end_cells)):
                        plt.text(j, i, str(gridworld[i, j]), 
                            ha='center', va='center', fontsize=12, color='black')
                        
    else:    # in this case print actions
        for i in range(gridworld.shape[0]):
            for j in range(gridworld.shape[1]):
                
                if (i, j) not in (forbidden_cells + end_cells):
                    print(f'arrow in {i},{j}')
                    action = int(gridworld[i, j])
                    
                    dx, dy = 0, 0
                    arrow_scale = 0.3
                    
                    if action == 0:      # up
                        dy = -arrow_scale
                    elif action == 1:    # down
                        dy = arrow_scale
                    elif action == 2:    # left
                        dx = -arrow_scale
                    elif action == 3:    # right
                        dx = arrow_scale
                    
                    plt.arrow(
                        j - dx*0.5,
                        i - dy*0.5,
                        dx,
                        dy,
                        head_width=0.25,
                        head_length=0.25,
                        fc='black',
                        ec='black',
                        linewidth=2,
                        length_includes_head=True,
                        zorder=5
                    )

            
    # color in green starting cells
    for start_cell in start_cells:
        row, col = start_cell
        color = 'green'
        rect = Rectangle((col - 0.5, row - 0.5), 1, 1, 
                     facecolor=color, edgecolor='none', alpha=0.8)
        plt.gca().add_patch(rect)
        
    # color in red forbidden cells
    for start_cell in forbidden_cells:
        row, col = start_cell
        color = 'red'
        rect = Rectangle((col - 0.5, row - 0.5), 1, 1, 
                     facecolor=color, edgecolor='none', alpha=0.8)
        plt.gca().add_patch(rect)

    # color in grey end cells
    for start_cell in end_cells:
        row, col = start_cell
        color = 'grey'
        rect = Rectangle((col - 0.5, row - 0.5), 1, 1, 
                     facecolor=color, edgecolor='none', alpha=0.8)
        plt.gca().add_patch(rect)

    # Create custom legend
    legend_elements = [
        Patch(facecolor='green', edgecolor='black', label='starting cell', alpha=0.8),
        Patch(facecolor='red', edgecolor='black', label='forbidden cell', alpha=0.8),
        Patch(facecolor='grey', edgecolor='black', label='terminal cell', alpha=0.8)
    ]
    
    plt.legend(handles=legend_elements, loc='upper left', bbox_to_anchor=(1, 1))
    
    
    plt.show()


def plot_actions(actions):
    arrows = {0: (0,1), 1:(0,-1), 2:(-1,0), 3: (1,0)}
    scale = 0.25
    fig, ax = plt.subplots(figsize=(6, 6))
    for r, row in enumerate(actions):
        for c, cell in enumerate(row):
            plt.arrow(c, 5-r, scale*arrows[cell][0], scale*arrows[cell][1], head_width=0.1)
    
    plt.show()
```

## 1: Basic Gridworld and standard solution
We want to define a gridworld with the following rules:
- an agent can move one step at a time in an adjacent cell in directions up, down, left and right
- all actions are always available, but if no cell is present (e.g. there is a wall) the agent will land on the same cell he is in
- the agent shall exit the gridworld in as few possible steps
- the score of the agent decreases of one point every time he takes a step

```python
# basic gridworld setup
l = 5      ### length of the gridworld edge (square)
gridworld = np.zeros((l,l))

### start and exit cells
starts = [(0,0)]
ends = [(l-1,l-1)]
forbiddens = []

### action list
actions = ['up','down','left','right']

plot_gridworld(gridworld,starts,forbiddens,ends,gradient = False,actions = [])

```

This is a very basic gridworld with a starting cell and an ending cell, with no forbidden cells in the middle. Let's try to solve this with standard RL approaches such as Monte Carlo

## 1.1: Monte Carlo Gridworld solution 
We will now use a MC approach to solve the gridworld. The initial setup for the action value function $Q(s,a)$ is a random initialization (uniform Q for each State/Action pair) and the policy $\pi(s)$ is arbitrary

```python
# initial setup
Q = np.zeros((l,l,len(actions)))   # state:(coor x, coor y), action: up,down,left,right


# setup policy randomly (uniform across all actions)
pi = np.ones((l,l,len(actions)))
for i in range(np.shape(pi)[0]):
    for j in range(np.shape(pi)[1]):
        norm_factor = np.sum(pi[i,j,:])
        pi[i,j,:] = pi[i,j,:] / norm_factor


# return saver
returns = []
```

```python
# maps name of an action into a numeric index
def action_index(action_name):
    if (action_name == 'up'):
        return 0
    elif(action_name == 'down'):
        return 1
    elif(action_name == 'left'):
        return 2
    elif(action_name == 'right'):
        return 3
    else:
        print('Error in action name!')
        return None

# evaluates the policy on n_episodes
def evaluate_policy(pi_eval, n_episodes, gamma):
    total_return = 0
    
    for _ in range(n_episodes):
        idx = np.random.choice(len(starts))
        state = np.array(starts[idx])
        
        G = 0
        discount = 1
        done = False
        steps = 0
        max_steps = 100
        
        while not done and steps < max_steps:
            action = pi_eval[state[0], state[1]]
            
            new_state = state.copy()
            
            if action == 0: new_state[1] -= 1
            elif action == 1: new_state[1] += 1
            elif action == 2: new_state[0] -= 1
            elif action == 3: new_state[0] += 1
            
            if (new_state[0] < 0 or new_state[0] >= l or 
                new_state[1] < 0 or new_state[1] >= l or
                any(np.array_equal(new_state, f) for f in forbiddens)):
                new_state = state.copy()
            
            state = new_state
            
            G += discount * (-1)
            discount *= gamma
            steps += 1
            
            if any(np.array_equal(state, e) for e in ends):
                done = True
        
        total_return += G
    
    return total_return / n_episodes
```

```python

returns_count = defaultdict(lambda: defaultdict(int))
total_iter = 10000
gamma = 0.8
epsilon = 0.05
steps = []
eval_every = int(total_iter/25)
max_steps = 100

# Then your loop works as-is
for iteration in range(total_iter):
    seen_s = []
    seen_a = []
    
    # Initial state
    idx = np.random.choice(len(starts))
    curr_s = np.array(starts[idx])
    
    finished = False
    
    # Generate episode
    step_count = 0
    while not finished and step_count < max_steps:
        curr_a = np.random.choice(actions, p=pi[curr_s[0], curr_s[1], :])
        seen_s.append(curr_s.copy())
        seen_a.append(curr_a)
        
        # Take action
        new_s = curr_s.copy()
        if curr_a == 'up':
            new_s[1] -= 1
        elif curr_a == 'down':
            new_s[1] += 1
        elif curr_a == 'left':
            new_s[0] -= 1
        elif curr_a == 'right':
            new_s[0] += 1
        
        # Check boundaries and forbidden states
        if (new_s[0] < 0 or new_s[0] >= l or 
            new_s[1] < 0 or new_s[1] >= l or
            any(np.array_equal(new_s, forb) for forb in forbiddens)):
            new_s = curr_s.copy()
        
        curr_s = new_s
        step_count +=1
        
        # Check terminal
        if any(np.array_equal(curr_s, end) for end in ends):
            finished = True
        

    # save num steps
    steps.append(len(seen_s))
    # Calculate returns (first-visit MC)
    G = 0
    visited = set()
    
    for t in range(len(seen_s) - 1, -1, -1):
        G = gamma * G - 1
        
        state_tuple = tuple(seen_s[t])
        action_idx = action_index(seen_a[t])
        
        if (state_tuple, action_idx) not in visited:
            visited.add((state_tuple, action_idx))
            
            # Update Q with incremental average
            count = returns_count[state_tuple][action_idx]
            Q[seen_s[t][0], seen_s[t][1], action_idx] = \
                (Q[seen_s[t][0], seen_s[t][1], action_idx] * count + G) / (count + 1)
            returns_count[state_tuple][action_idx] += 1
            
            # Update policy (epsilon-greedy)
            A_star = np.argmax(Q[seen_s[t][0], seen_s[t][1], :])
            for a in range(len(actions)):
                if a == A_star:
                    pi[seen_s[t][0], seen_s[t][1], a] = 1 - epsilon + epsilon / len(actions)
                else:
                    pi[seen_s[t][0], seen_s[t][1], a] = epsilon / len(actions)
                    
    #print(np.sum(np.abs(Q)))
    steps.append(step_count)
    #if iteration % eval_every == 0:
    #    pi_eval = np.argmax(Q,axis = 2)
    #    #print(pi_eval)
    #    avg_return = evaluate_policy(pi_eval,100,gamma)
    #    performances.append(avg_return)
```

```python
Q_star = np.zeros((l,l))
A_star = np.zeros((l,l))
for x_coor in range(np.shape(Q)[0]):
    for y_coor in range(np.shape(Q)[1]):
        Q_star[x_coor,y_coor] = round(np.max(Q[x_coor,y_coor,:]),3)
        A_star[x_coor,y_coor] = np.argmax(Q[x_coor,y_coor,:])
        
```

```python
plot_gridworld(Q_star,starts,forbiddens,ends, gradient = True, actions = [])
```

```python
pi[2,2,:]
```

```python
# evaluation of accuracy of MC
best_path_idx = [1. if a == 8 else 0. for a in steps]
best_path_idx = np.cumsum(best_path_idx)
for idx in range(len(best_path_idx)):
    best_path_idx[idx] = best_path_idx[idx] / (idx+1)

stopping_t = 2000
plt.plot(np.arange(0,len(best_path_idx[:stopping_t]),1),best_path_idx[:stopping_t], label = 'standard MC')
plt.xlabel('MC iteration')
plt.ylabel('% of best path choice')
plt.grid(alpha = 0.6)
plt.legend()
plt.show()
```

## 2: A more complex gridworld example

```python

returns_count = defaultdict(lambda: defaultdict(int))
total_iter = 1000000
gamma = 1.
epsilon = 0.05
steps = []
eval_every = int(total_iter/25)
max_steps = 1000

# Then your loop works as-is
for iteration in range(total_iter):
    seen_s = []
    seen_a = []
    
    # Initial state
    idx = np.random.choice(len(starts))
    curr_s = np.array(starts[idx])
    
    finished = False
    
    # Generate episode
    step_count = 0
    while not finished and step_count < max_steps:
        curr_a = np.random.choice(actions, p=pi[curr_s[0], curr_s[1], :])
        seen_s.append(curr_s.copy())
        seen_a.append(curr_a)
        
        # Take action
        new_s = curr_s.copy()
        if curr_a == 'up':
            new_s[1] -= 1
        elif curr_a == 'down':
            new_s[1] += 1
        elif curr_a == 'left':
            new_s[0] -= 1
        elif curr_a == 'right':
            new_s[0] += 1
        
        # Check boundaries and forbidden states
        if (new_s[0] < 0 or new_s[0] >= l or 
            new_s[1] < 0 or new_s[1] >= l or
            any(np.array_equal(new_s, forb) for forb in forbiddens)):
            new_s = curr_s.copy()
        
        curr_s = new_s
        step_count +=1
        
        # Check terminal
        if any(np.array_equal(curr_s, end) for end in ends):
            finished = True
        

    # save num steps
    steps.append(len(seen_s))
    # Calculate returns (first-visit MC)
    G = 0
    visited = set()
    
    for t in range(len(seen_s) - 1, -1, -1):
        G = gamma * G - 1
        
        state_tuple = tuple(seen_s[t])
        action_idx = action_index(seen_a[t])
        
        if (state_tuple, action_idx) not in visited:
            visited.add((state_tuple, action_idx))
            
            # Update Q with incremental average
            count = returns_count[state_tuple][action_idx]
            Q[seen_s[t][0], seen_s[t][1], action_idx] = \
                (Q[seen_s[t][0], seen_s[t][1], action_idx] * count + G) / (count + 1)
            returns_count[state_tuple][action_idx] += 1
            
            # Update policy (epsilon-greedy)
            A_star = np.argmax(Q[seen_s[t][0], seen_s[t][1], :])
            for a in range(len(actions)):
                if a == A_star:
                    pi[seen_s[t][0], seen_s[t][1], a] = 1 - epsilon + epsilon / len(actions)
                else:
                    pi[seen_s[t][0], seen_s[t][1], a] = epsilon / len(actions)
                    
    #print(np.sum(np.abs(Q)))
    steps.append(step_count)
    #if iteration % eval_every == 0:
    #    pi_eval = np.argmax(Q,axis = 2)
    #    #print(pi_eval)
    #    avg_return = evaluate_policy(pi_eval,100,gamma)
    #    performances.append(avg_return)
```

```python
Q_star = np.zeros((l,l))
A_star = np.zeros((l,l))
for x_coor in range(np.shape(Q)[0]):
    for y_coor in range(np.shape(Q)[1]):
        Q_star[x_coor,y_coor] = round(np.max(Q[x_coor,y_coor,:]),1)
        A_star[x_coor,y_coor] = np.argmax(Q[x_coor,y_coor,:])
        

plot_gridworld(Q_star,starts,forbiddens,ends, gradient = True, actions = [])
```

```python
# evaluation of accuracy of MC
best_steps = 14
best_path_idx = [1. if a == best_steps else 0. for a in steps]
best_path_idx = np.cumsum(best_path_idx)
for idx in range(len(best_path_idx)):
    best_path_idx[idx] = best_path_idx[idx] / (idx+1)


plt.plot(np.arange(0,len(best_path_idx[:]),1),best_path_idx[:], label = 'standard MC')
plt.xlabel('MC iteration')
plt.ylabel('% of best path choice')
plt.grid(alpha = 0.6)
plt.legend()
plt.show()
```
