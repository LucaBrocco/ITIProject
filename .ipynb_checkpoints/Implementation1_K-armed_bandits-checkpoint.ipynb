## Formalization of the problem
We have an agent aiming to solve a k-armed bandits problem. We can imagine the problem as a slot machine with k levers. Pulling one of them gives out the highest amount of cash. We want our agent to be able to get the highest amount of rewards on the long run. To encode this structure, I am going to code a setup such as the one presented in 10-armed Testbed: I am going to extract 10 numbers from a normal distribution. These numbers will represent the average Reward given by each slot machine, according to normal distributions with fixed variance. It is notable that the problem can be generalized to any kind of distribution.

```python
import numpy as np
import matplotlib.pyplot as plt

np.random.seed(3110) # set seed to replicate results

# parameters of our "hidden distribution" of prizes
n_levers = 10
mean = 0
std = 1
q_star = np.random.normal(loc = mean, scale = std, size = n_levers) # true value of the means 

plt.errorbar(np.arange(1,len(q_star)+1,1),q_star,yerr = np.tile(1.,len(q_star)), marker = 'o', ls = 'none')
plt.ylabel('q* distribution')
plt.xlabel('k')
plt.show()
```

The plot above shows the distribution of the rewards. There is clearly a preferrable choice (the lever whose q* has an higher average), but the agent has to be able to identify it first. To achieve this, there are several algorithms:

## Greedy choice
We need the agent to be able to estimate how good each action can be. For this, the definition of estimate of Q-value can be used:

$Q_t(A) = \frac{\sum{R}}{N}$

Which corresponds to the sum of the rewards got from picking action A over the number of times A was picked.
For the law of large numbers this quantity converges to $q*$ for a large amount of trials. This enables the agent, with sufficient trial and error, to be able to identify correctly which action is the best one to take (identified as **Greedy action**). A way to estimate this quantity for each timestep $t$ (which in context of k-armed bandits corresponds to an episode) is to compute as follows:

$Q_{t+1} = Q_t + \frac{(R_t - Q_t)}{n}$

This expression allows for easy storing of the data we need: no risk of overflow is met with such an update.

We would like the agent to be able to exploit the greedy action to get the highest amount of rewards. A naif approach would be an algorithm that tries once all k levers and then sticks to the action found to give the highest reward amount.

```python
def greedy_algorithm_with_first_push(Q,t):
    # try out all possibilities
    if t < len(Q):
        # action selection
        a_idx = t
    # if he has already ran once all the levers, just pick the best one
    else:
        a_idx = np.argmax(Q)
    # return the chosen lever
    return a_idx
```

```python
def reward_generator(a_idx, q_star, std):
    # for a given choice, the bandit give back a reward extracted from a gaus distrib.
    R = np.random.normal(loc = q_star[a_idx],scale = std) # if size is not specified a scalar is returned
    return R

def update_Q(Q,N,a_idx,R):
    # update the estimate of Q according to what found
    Q[a_idx] = Q[a_idx] + (R - Q[a_idx]) / N[a_idx]
    return Q

def optimal_action_metric(actions, best_action):
    num = np.cumsum(actions == best_action) 
    den = np.arange(1,len(actions)+1,1)
    return num/den
```

```python
# initialization of Nruns
timesteps = 1000 # for how long the agent must play the bandit
metric_saver_fp = np.zeros(timesteps)
n_runs = 2000

for i in (range(n_runs)):
    # initialization of algorithm (re-extract the bandits every time)
    q_star = np.random.normal(loc = mean, scale = std, size = n_levers) # true value of the means 
    Q = np.zeros(n_levers)
    N = np.zeros(n_levers)
    
    best_action = np.argmax(q_star)
    actions = np.zeros(timesteps) # save the actions
    # loop
    for t in (range(timesteps)):
        a_idx = greedy_algorithm_with_first_push(Q,t)   # pick an action
        N[a_idx] += 1                                   # increase the counter for that action
        R = reward_generator(a_idx,q_star,std)          # generate the reward
        Q = update_Q(Q,N,a_idx,R)                       # update Q
        actions[t] = a_idx
    
    # convert actions array in cumulative times of best action taken
    metric = optimal_action_metric(actions,best_action)
    metric_saver_fp = metric_saver + (metric - metric_saver_fp)/(i+1)
    
plt.plot(np.arange(0,timesteps,1),metric_saver_fp)
plt.xlabel('time t')
plt.ylabel('Fraction for optimal action')
plt.show()

    
```

The above graph shows how the **greedy algorithm** performs with pushing each lever one time and proceeding from then on with the lever found to be the best one. One could wonder what would happen if instead the algorithm just picks the highest value lever and breaks ties with uniform distribution:

```python
def silly_greedy_algorithm(Q,t):
    # find out the value of the max
    a_idx = np.argmax(Q)
    # check if there are more than one == to max
    n_max = np.where(Q == Q[a_idx])[0]
    # if there is only one, that's our man!
    if len(n_max == 1):
        return a_idx
    # if there are ties, we should pick one randomly
    else:
        a_idx = int(len(n_max) * np.random.uniform(0,1))
        return a_idx
    
```

As you can see, just pressing once each lever allows the agent to have a significantly better performance over time. This is thanks to the **exploration** it has performed at the start. The agent sacrifices some rewards initially to maximize long term ones. We would like to optimize this **exploration-exploitation** tradeoff even more, by identifying as soon as possible the best lever.

## $\epsilon$-greedy

An algorithm that allows us to achieve better performances by exploiting this tradeoff is called **$\epsilon$-greedy**. We define a small quantity $\epsilon$ that allows the agent some exploration: he will take with probability $1-\epsilon$ the greedy action, while with probability $\epsilon$ will take another action at random. This way the agent will mostly pick the actions known to be best at that point in time, but will still explore, allowing himself to find out if another action can be better. 

```python
def epsilon_greedy_algorithm(Q,epsilon):
    # find out the value of the max
    a_idx = np.argmax(Q)
    # extract a number to decide what to do
    dec = np.random.uniform(0,1)
    
    # greedy choice
    if (dec > epsilon):
        # check if there are more than one == to max
        n_max = np.where(Q == Q[a_idx])[0]
        # if there is only one, that's our man!
        if len(n_max == 1):
            return a_idx
        # if there are ties, we should pick one randomly
        else:
            a_idx = int(len(n_max) * np.random.uniform(0,1))
            return a_idx
    # non-greedy choice
    else:
        # possible choices
        pos = np.arange(0,len(Q),1)
        pos = pos[pos != a_idx] # remove the argmax
        a_idx = pos[int(len(pos)*np.random.uniform(0,1))] # pick one random
        return a_idx
        
    
```

```python
# initialization of Nruns
timesteps = 1000 # for how long the agent must play the bandit
n_runs = 2000
epsilons = [0.1,0.01]

for epsilon in epsilons:
    metric_saver_eg = np.zeros(timesteps)
    for i in (range(n_runs)):
        # initialization of algorithm (re-extract the bandits every time)
        q_star = np.random.normal(loc = mean, scale = std, size = n_levers) # true value of the means 
        Q = np.zeros(n_levers)
        N = np.zeros(n_levers)
        
        best_action = np.argmax(q_star)
        actions = np.zeros(timesteps) # save the actions
        # loop
        for t in (range(timesteps)):
            a_idx = epsilon_greedy_algorithm(Q,epsilon)   # pick an action
            N[a_idx] += 1                                   # increase the counter for that action
            R = reward_generator(a_idx,q_star,std)          # generate the reward
            Q = update_Q(Q,N,a_idx,R)                       # update Q
            actions[t] = a_idx
        
        # convert actions array in cumulative times of best action taken
        metric = optimal_action_metric(actions,best_action)
        metric_saver_eg = metric_saver_eg + (metric - metric_saver_eg)/(i+1)
    
    plt.plot(np.arange(0,timesteps,1),metric_saver_eg, label = f'$\epsilon = {epsilon}$')
plt.plot(np.arange(0,timesteps,1),metric_saver_g, label = 'greedy')
plt.xlabel('time t')
plt.ylabel('Fraction for optimal action')
plt.legend()
plt.show()

    
```

In the above plot I wanted to check whether $\epsilon$-greedy would overcome other approaches. As $\epsilon$ decreases, the increase of the accuracy with respect to time is slower, but it will approach $1-\epsilon$ at the end. The algorithm will still fail a fraction $\epsilon$ of times at convergence, since it still picks non-greedy actions at random, even if he has discovered which is the best action. To achieve the best performance, the value of $\epsilon$ should be adjusted according to runtime $t$.

## Optimistic Initial Values

We can help the algorithm understand which is the best action by providing "optimistic" initial values. This enhances exploration by the algorithm, since he finds himself thinking that some actions will yield even higher rewards than they will ever could. To apply this, we need some prior knowledge about the problem. In this context, we know that our $q^*$ distribution is constructed out from gaussian values extracted with $\mu=0$ and $\sigma=1$. Optimistic initial values are then in $3-4$ range


```python
# initialization of Nruns
timesteps = 1000 # for how long the agent must play the bandit
n_runs = 2000
epsilons = [0.1,0.01]
Q_zeros = [0,3.5]

for epsilon in epsilons:
    for q_zero in Q_zeros:
        metric_saver_egoiv = np.zeros(timesteps)
        for i in (range(n_runs)):
            # initialization of algorithm (re-extract the bandits every time)
            q_star = np.random.normal(loc = mean, scale = std, size = n_levers) # true value of the means 

            if q_zero == 0:
                Q = np.zeros(n_levers)
            else:
                Q = np.ones(n_levers)*q_zero # optimistic initialization
            N = np.zeros(n_levers)
            
            best_action = np.argmax(q_star)
            actions = np.zeros(timesteps) # save the actions
            # loop
            for t in (range(timesteps)):
                a_idx = epsilon_greedy_algorithm(Q,epsilon)   # pick an action
                N[a_idx] += 1                                   # increase the counter for that action
                R = reward_generator(a_idx,q_star,std)          # generate the reward
                Q = update_Q(Q,N,a_idx,R)                       # update Q
                actions[t] = a_idx
            
            # convert actions array in cumulative times of best action taken
            metric = optimal_action_metric(actions,best_action)
            metric_saver_egoiv = metric_saver_egoiv + (metric - metric_saver_egoiv)/(i+1)
        
        plt.plot(np.arange(0,timesteps,1),metric_saver_egoiv, label = f'$\epsilon = {epsilon}$, $Q_0 = {q_zero}$')
#plt.plot(np.arange(0,timesteps,1),metric_saver_g, label = 'greedy')
plt.xlabel('time t')
plt.ylabel('Fraction for optimal action')
plt.legend()
plt.show()

    
```

From the above plot we see that, in general, having Optimistic Initial Values improves the performance, at parity of $\epsilon$ value

## Upper-Confidence bound

We would like to do better than picking the non-greedy action with uniform probability. In general, we would prefer actions that have _chance_ of performing better. Trying them out, we could discover that they lead to higher rewards, or maybe lower. In the latter case, we would have spent one timestep not getting high rewards, but we would have gained information. This can be achieved computing $A_t$ considering the confidence we have over each single action: 

```python
def UCB_algorithm(Q,c,N,t):
    a_idx = np.argmax(Q + c*np.sqrt(np.log(t)/N)) # argmax of this is the desired index
    # check for ties and break them randomly
    n_max = np.where(Q == Q[a_idx])[0]
    # if there is only one, that's our man!
    if len(n_max == 1):
        return a_idx
    # if there are ties, we should pick one randomly
    else:
        a_idx = int(len(n_max) * np.random.uniform(0,1))
        return a_idx
```

```python
# initialization of Nruns
timesteps = 1000 # for how long the agent must play the bandit
n_runs = 2000
c_s = [0.01,0.1,1.,2.,4.]


for c in c_s:
    metric_saver_UCB = np.zeros(timesteps)
    for i in (range(n_runs)):
        # initialization of algorithm (re-extract the bandits every time)
        q_star = np.random.normal(loc = mean, scale = std, size = n_levers) # true value of the means 

        Q = np.zeros(n_levers)
        N = np.ones(n_levers)   # avoid division by 0
            
        best_action = np.argmax(q_star)
        actions = np.zeros(timesteps) # save the actions
        # loop
        for t in (range(timesteps)):
            a_idx = UCB_algorithm(Q,c,N,t)                  # pick an action
            N[a_idx] += 1                                   # increase the counter for that action
            R = reward_generator(a_idx,q_star,std)          # generate the reward
            Q = update_Q(Q,N,a_idx,R)                       # update Q
            actions[t] = a_idx
            
        # convert actions array in cumulative times of best action taken
        metric = optimal_action_metric(actions,best_action)
        metric_saver_UCB = metric_saver_UCB + (metric - metric_saver_UCB)/(i+1)
    plt.plot(np.arange(0,timesteps,1),metric_saver_UCB, label = f'UCB $c = {c}$')
        
plt.plot(np.arange(0,timesteps,1),metric_saver_egoiv, label = f'$\epsilon = {epsilon}$, $Q_0 = {q_zero}$')
#plt.plot(np.arange(0,timesteps,1),metric_saver_g, label = 'greedy')
plt.xlabel('time t')
plt.ylabel('Fraction for optimal action')
plt.legend()
plt.show()

    
```
